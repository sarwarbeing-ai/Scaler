## Probability Distribution

#### Q. What are probability distributions?
- It describes how a random variable is **distributed** among the range of values it contains.

- It also tells how likely it is for random variable to a take **particular value.**

- The likelihood of a random variable taking a specific value at a point is called the **probability density** at that point.

#### Q. What are some different types of probability distributions?
Recall the Statistics class where we studied about the following:-
1. PDF (Probability Density Function)
2. CDF (Cummulative Distribution Function)
3. PMF (Probability Mass Function)
4. PPF (Percent Point Function)

Let's recall PDF and CDF, and we will see their functionality in relation to Naive Bayes.

#### Q.When should we use Distributions?
- Used for doing **EDA** on the dataset
- Distributions are "simple models" of natural phenomenon
- They tell a lot of properties of a random variable like - mean, std, skewness etc

## CDF (Cummulative Distribution Function)



#### Q. What does the CDF signify?
The CDF of observations $X = x_1, x_2, x_3, ..., x_n$ give the probability of any value in the random variable X being less than or equal to a particular value 'x'. Here, X is a **real valued random variable**.

X can be either a continuous random variable, or a discrete random variable.

The CDF is the representation of the probabilities of each outcome with the outcomes previous to it.

#### Q. Give example of a case where we would use CDF
For e.g. if we want the probability of students getting marks less than 70 in an exam.

#### Q. How is CDF calculated?
Recall that CDF is the probability that our random variable X will have a value less than or equal to a sample point x

Hence, The formula to calculate CDF is:-
$F_X (x) = P(X <= x) $

---
#### To give an example of CDF
- Consider the throw of a dice.
 - We can only get {1, 2, 3, 4, 5, 6} as the possible outcomes.
 - And we know the probablity of getting any one of these values is 1/6.

- Recall that The CDF is the representation of the probabilities of each outcome with the outcomes previous to it.
- On plotting the CDF for this event, we get:
 - The values of each probability, i.e., P(X<=1), P(X<=2), ........, P(X<=6) are plotted on y axis against the values on the x axis.
 - The points are joined and a straight line is obtained.
 - The line is non decreasing and has a maximum value of 1

- It is non decreasing and right continuous

## PDF (Probability Density Function)



#### Q. What does the PDF signify?
- PDF is used to express the **frequency distributions of continuous random variable**.
- E.g. the heights of humans, marks of students.
- It maps the value of the random variable to their corresponding **probability densities**.
- The y-axis does not signifies the probabilities, **area under the curve** does.

#### Q. How do we calculate the PDF value?
- Let $\mu$ be the mean of the data 
- Let $$ be the standard deviation of the data
- Then the PDF for a random variable x is calculated as:
$f(x) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{\frac{1}{2}(\frac{x-\mu}{\sigma})^2}$

#### Q. How does the value of $$ and $$ affect the PDF?
- The value of mean ($$) determines where the distribution will be **centered**.
- The value of standard deviation ($^2$) determines how the distribution is spread 
- As you move away from the mean, probability density decreases **exponentially**.

#### Q. Is there any relation between PDF and CDF?
- Consider a continuous random variable X.
- We obtain it's CDF as F(X). We find that it is absolutely continuous. 
- If we were to **differentiate** this function F(X), we would obtain PDF for X.

$PDF(X)=\frac{CDF(x)}{dx}=CDF(X)$

#### Q. What are some properties of PDF?
1. The probability density function is non-negative for all the possible values, i.e. $f(x) 0$, for all x.
2. The area between the density curve and horizontal X-axis is equal to 1, i.e. $\int_{-\infty}^{\infty}f(x).dx = 1$
3. Due to the property of continuous random variables, the density function curve is **continued** for all over the given range.

## Naive Bayes (NB) on non text data


#### Q. What if our data is real valued (non-text data), will N.B. still work?
**Yes.**

Essentially, in order to make our Naive Bayes classifier work, we need to be able to compute **class priors** and **likelihood** values. If we can do that, then N.B. would work.

---
#### Let's understand this with an example
- Consider we have real valued data related to an individual's health, and we wish to classify the person as diabetic or non-diabetic.
- Here the feature values $f_j$ are **continuous randomn variables**.

#### Q. How would we calculate the class prior values for real valued data in this example?

The calculation for class priors remain the **same**. 
- Let the total number of individuals in the data be $n$
- Let the number of individuals with with diabetes (class $y=1$) be $n_1$.
- Similarily, Let the number of samples without diabetes (class $y=0$) be $n_0$.
- Then, class priors are calculated as:-
 - $P(y=1) = \frac{n_1}{n}$
 - $P(y=0) = \frac{n_0}{n}$

---
Suppose in our data, the value of jth feature $f_j$ for the ith person is 10.2
#### Q. How would we calculate the likelihood values for real valued data in this example?
Calculation of the likelihood values for real valued data is **slightly different** than what we studied for text data. There is a slight twist!

- Now when we say that we need to calculate the likelihood of feature j $f_j$ for ith person, we actually have to calculate $P(f_j =10.2|y=1)$
- In English, we call this as "the probability that the value of feature j is 10.2 given that the individual is diabetic."

- Similarily, we can calculate $P(f_j =10.2|y=0)$

Now, just like before, we compare these two values $P(f_j =10.2|y=1)$ and $P(f_j =10.2|y=0)$, and return the higher value as our result.

### Q.How do you think we can solve for the likelihood values here?
We need to calculate the likelihoods as $P(f_j =10.2|y=1)$ where 10.2 is just some random real value we assumed.
- We know that feature $f_j$ is a continuous random variable.
- So essentially what we wish to calculate (i.e. **relative likelihood**), is typically known as **Conditional PDF of random variable $f_j$** i.e. $P(f_j)=Conditional \ PDF \ of \ f_j$

Lets explain this further.

- We obtain the PDF for these values.
- So essentially, by doing so, we obtain the **PDF of feature $f_j$ conditioned on class 1** i.e. the **conditional PDF**

## Gaussian Naive Bayes



### Q. What if the PDF we obtain has a Gaussian shape?
**If all conditional PDFs we get are Gaussian, then it is what we call Gaussian Naive Bayes**

#### Q. What do we mean by Gaussian shape/nature?
By Gaussian nature, we mean that the probability distribution is **symmetric about the mean**, showing that data near the mean are more frequent in occurrence than data far from the mean.

#### Q. Does this mean, that for it to be Gaussian NB, all the features have to exhibit Gaussian nature?
No. 

For it to be Gaussian NB, the **variables themselves may not be Gaussian!** 

But if the **conditional PDFs** (i.e. PDF of feature $f_j$ conditioned on y=1 or y=0) are Gaussian, then we have Gaussian N.B.

gaus.png

### Q. How do we know if the conditional PDFs are in fact Gaussian?
As you may recall, we have several tests to check for the same:-
- QQ plots
- Shapiro-Wilk test
- Anderson Darling test
- KS test

### Q. What do we do during the **training time** in Gaussian NB?
- Calculate the **class priors**
 - i.e. $P(y=0)$ and $P(y=1)$

- Now we need to calculate our **likelihoods**
 - Instead of likelihoods, we'll get the whole distribution itself.
 - So, for each feature $f_j$ conditioned on class i.e. $f_j|y=1$
 - Suppose, we find out that this is **Normal (Gaussian) distribution** i.e. 
   - $f_j|y=1$ can be represented in the normal form as $ N(_{j1}, _{j1})$ and 
   - $f_j|y=0$ can be represented in the normal form as $N(_{j0}, _{j0})$ <br> where $$ -> Mean <br>$$ -> Standard deviation
 - Then, from the conditional PDF plot, We can get the $$ and $$ values.
 
- Also, we need to **check** that the data actually is Gaussian using the tests we mentioned.

As we were exploring using NB on non-text data, let's go back to our diabetes example. 

At test time, we get a new data point, say $x_q$, consisting of d features.

### Q. What happens during the **testing time** in Gaussian NB?

- What do we **want to compute?**
 - We want to compute that given an individual's information as features, let the row be $x_q$, what is the probability that the person is diabetic (class=1) (i.e. $P(y=1|x_q)$)
 - This as we've seen many times now, is calculated as $P(y=1|x_q)  \ P(y=1) *_{i=1}^d P(f_i|y=1)$

---
#### Q. How do we obtain the class prior values?
The class prior values were already calculated during training.

---
Suppose there is a feature $f_j$ in the query point $x_q$, whose feature value is say, 1.2

#### Q. How do we obtain the likelihood value?
- The likelihood term $P(f_i|y=1)$ actually calculates the probability that value of feature j $f_j$ is equal to 1.2, given the person is diabetic (i.e. class=1),  (i.e. $P(f_j=1.2|y=1)$)
 - This **feature conditioned at class 1** (i.e. $f_j|y=1$), we already know is **Gaussian distributed**.
 - We can get this value $P(f_j=1.2|y=1)$ from the corresponding height of the **Conditional PDF curve for $f_j|y=1$** at x=1.2


- Similarily, we can obtain the probability that value of feature j $f_j$ is equal to 1.2, given the person is not diabetic (i.e. class=0), i.e. the likelihood term $P(f_i=1.2|y=0)$
 - This **feature conditioned at class 0** (i.e. $f_j|y=0$), we already know is **Gaussian distributed**.
 - We can get this value $P(f_j=1.2|y=0)$ from the corresponding height of the **Conditional PDF curve for $f_j|y=0$** at x=1.2

---
Plugging the class prior and likelihood values into the equation, we obtain the value of $P(y=1|x_q)$ and $P(y=0|x_q)$.

Now, we just compare these 2 values, and based on the greater value, predict the person as diabetic or non-diabetic.



postRead_Gauss.jpg


So, when we have real valued features, we're essentially **fitting a probability distribution to the conditional probabilities**.


---
So, to recap Gaussian N.B.
- At **training time** we're **fitting Gaussian distribution to each of these conditional random variables**. 

- And at **runtime** (testing time), we compute these **conditional probabilities using the normal distribution** (i.e. $N(_{j1}, _{j1})$). We plug it in, and everything will flow!
