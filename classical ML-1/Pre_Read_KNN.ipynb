### Euclidean Distance

- The most basic distance measure can be done using **Euclidean Distance**


- Which is Algebrically just the sum of the square difference between each of the d-dimensional point of $x_1$ with $x_2$ and taking a square root of it.
    - Euclidean $(x_1,x_2) = [ âˆ‘_{j=1}^{j=d} {{ (x _{1j} - x _{2j} )}^2] }^{\frac{1}{2}} $ 

- Geometrically, Euclidean distance is the length of the Hypothesis of the right angle triangle drawn between two points $x_1$ and $x_2$.

Note: Euclidean distance is also called flying distance, since it is the shortest distance between two points in  d-dimensional space.
